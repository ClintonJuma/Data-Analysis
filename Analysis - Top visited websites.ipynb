### importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### importing dataset

```python
data = pd.read_csv("Top visited websites.csv")
data
```

### Basic information

```python
#shape
data.shape
```

```python
data.info()
```

```python
data.columns.str.split(",").value_counts()
```

```python
#dropping column Unnamed
data.drop(labels='Unnamed: 0', axis=1)
```

```python
#checking for null values
data.isnull().sum()
```

```python
data["Category"].nunique()
```

```python
#checking category and count
data["Category"].str.split(",").value_counts()
```

```python
#checking country and count
data["Principal country/territory"].str.split(",").value_counts()
```

```python
#grouping by country
by_country = data.groupby(["Principal country/territory"])["Principal country/territory"].count().reset_index(name="count")
by_country = by_country.set_index("count")
by_country = by_country.sort_values("count",ascending=False)
by_country 
```

```python
#grouping by Category
by_Category = data.groupby(["Category"])["Category"].count().reset_index(name="count")
by_Category = by_Category.set_index("count")
by_Category = by_Category.sort_values("count",ascending=False)
by_Category
```

```python
#grouping by Category and Country
by_Category_Site = data.groupby(["Category","Principal country/territory"])["Category"].count().reset_index(name="count")
by_Category_Site = by_Category_Site.sort_values("count",ascending=False)
by_Category_Site = by_Category_Site.set_index("count")
by_Category_Site
```

```python
Top_by_Category_Site = by_Category_Site.query("`count`>1")
Top_by_Category_Site
```

```python
#describe data
data.describe(exclude="number")
```

## visualization

```python
#category plot
plt.figure(figsize = (10,7))
sns.countplot(y=data["Category"])
plt.xticks(rotation=75)
plt.title("Category Count")
```

More than 90% of the websites were of category, Computers Electronics and Technology > Social Media Networks followed by category News and Media publishers at around 80% and adult category at 60%

```python
#country plot
plt.figure(figsize = (10,7))
sns.countplot(x=data["Principal country/territory"])
plt.xticks(rotation=75)
plt.title(" Count")
plt.title("Country count")
```

Most websites were from United states at about 90%,followed by Russia at 30% and china at about 15%

```python
#Category by Country
plt.figure(figsize = (10,7))
plt.title("Category vs Country")
sns.barplot(x=Top_by_Category_Site.reset_index()["Category"],y=Top_by_Category_Site.reset_index()["count"],
           hue=Top_by_Category_Site.reset_index()["Principal country/territory"],ci=None)
plt.xticks(rotation=90)
```

In the top category by country,United states had the most category of websites in the Top,Russia had Computers Electronics and Technology > Social Media Networks and Cyprus had adult category
