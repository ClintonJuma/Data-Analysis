```python
#grouping based on Attempted suicide
Drunk = df.groupby(by = ["Country", "Year", "Age Group","Bullied","Really_Get_Drunk", "Use_Marijuana", "Had_sexual_relation", "Smoke_cig_currently", "Attempted_suicide"])["Use_Marijuana"].count().reset_index(name = "count")
Drunk = Drunk.sort_values("Attempted_suicide", ascending = False)
Drunk = Drunk.set_index("Attempted_suicide")
Drunk
```

```python
#grouping based on Bullying 
Fight = df.groupby(by= ["Country","Bullied", "Year", "Age Group","Had_fights", "Got_Seriously_injured", "Attempted_suicide"])["Bullied"].count().reset_index(name= "count")
Fight = Fight.sort_values("Bullied", ascending = False)
Fight = Fight.set_index("Country")
Fight
```

```python
TopSuicide = Drunk.query("`Attempted_suicide` >=30")
TopSuicide
```

```python
TopBully = Fight.query("`Bullied` >= 60")
TopBully
```
