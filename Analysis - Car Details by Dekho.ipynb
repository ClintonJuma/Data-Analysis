# Group by:

```python
df
```

```python
#group by tranmission, name, year, fuel and price
by_transmission = df.groupby(by= ["transmission", "name", "year", "selling_price", "fuel"])["name"].count().reset_index(name="count")
by_transmission = by_transmission.set_index("transmission")
by_transmission = by_transmission.sort_values("count", ascending = False)
by_transmission
```

```python
#grouping by fuel, name and tranmission
by_fuel = df.groupby(by = ["fuel", "name", "transmission"])["name"].count().reset_index(name="count")
by_fuel = by_fuel.set_index("fuel")
by_fuel = by_fuel.sort_values("count", ascending = False)
by_fuel
```

```python
#grouping by price, name, year and seller_type
by_price = df.groupby(by = ["name","year","selling_price", "seller_type", "fuel"])["year"].count().reset_index(name="count")
by_price = by_price.sort_values("selling_price", ascending = False)
by_price = by_price.set_index("selling_price")
by_price
```

```python
by_km_driven = df.groupby(by = ["name",  "km_driven", "selling_price", "fuel"])["name"].count().reset_index(name= "count")
by_km_driven = by_km_driven.set_index("km_driven")
by_km_driven = by_km_driven.sort_values("km_driven", ascending = False)
by_km_driven
```

```python
#Queries
transmission_query  = by_transmission.query("`count`> 10")
transmission_query
```

```python
#query of fuel group with more than 30 cars
fuel_query = by_fuel.query("`count` > 30")
fuel_query
```

```python
#query of cars above  500000
price_query = by_price.query("`selling_price`>500000")
price_query 
```

# Visualization

```python
sns.barplot(x = transmission_query.reset_index()["name"], y = transmission_query.reset_index()["count"],
            hue = transmission_query.reset_index()["transmission"])
plt.xticks(rotation=90)
plt.legend(ncol=10, loc="upper center")
plt.title("name vs count")
```

```python
sns.barplot(x = fuel_query.reset_index()["fuel"], y = fuel_query.reset_index()["count"])
plt.title("fuel vs count")
```

```python
sns.barplot(x = price_query.reset_index()["seller_type"], y = price_query.reset_index()["selling_price"], hue = price_query.reset_index()["fuel"])
plt.xticks(rotation = 65)
plt.title("seller type vs seller  price")
```
