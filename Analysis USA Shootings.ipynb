```python
#changing date column to datetime datatype
df["date"] = pd.to_datetime(df["date"])
df.info()
```

```python
#checking for duplicates
df.drop_duplicates(inplace=True)
df.info()
```

```python
#adding year, month, day columns
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month_name()
df["day"] = df["date"].dt.day_name()
df.info()
```

```python
#group by armed and race
armed_race=df.groupby(["armed", "race"])["race"].count().reset_index(name="count")
armed_race
```

```python
#group by armed and signs_of_mental_illness
armed_and_ill=df.groupby(["armed", "signs_of_mental_illness"])["armed"].count().reset_index(name="show_count")
armed_and_ill
```

```python
#group by armed and arms_category
ammo_type=df.groupby(["armed", "arms_category"])["armed"].count().reset_index(name="Count")
ammo_type
```

most suspects had Taser which is an electrical type of arm

```python
ammo_type["Count"].mean()
```

```python
#query for top arms
top_ammo=ammo_type.query("`Count` >=55")
top_ammo
```

```python
#group by threat level and arm category
ammo_threat = df.groupby(["threat_level", "arms_category"])["threat_level"].count().reset_index(name= "count")
ammo_threat
```

```python

ammo_threat["count"].mean()
```

```python
#query for top arms with the most threat
top_threat=ammo_threat.query("`count` >=169")
top_threat=top_threat.set_index("count")
top_threat
```

Guns and sharp objects  had the most threat, at level, Attack

```python
#grouping by threat level and race
threat_race=df.groupby(["threat_level", "race"])["race"].count().reset_index(name="count")
threat_race
```

```python
threat_race["count"].mean()
```

```python
#query for top threatful races
topthreat_race=threat_race.query("`count` >=169")
topthreat_race=topthreat_race.set_index("count")
topthreat_race
```

The white race  had the most threat, followed by Black then the Hispanic

```python
#grouped by race and arm_caterogy
race_ammo_category=df.groupby(["race", "arms_category"])["arms_category"].count().reset_index(name="count")
race_ammo_category
```

Whites armed with guns were the majority

```python

race_ammo_categoty=race_ammo_category.sort_values(by=["count"], ascending=True)
race_ammo_category.tail(5)
```

```python
#query for top race and arm
toprace_ammo=race_ammo_category.query("`count` >=84")
toprace_ammo

```

Whites with guns were the most, followed by blacks with guns and Hispanics with guns

```python
df.corr()
```

```python
df.plot(kind = "scatter", x  ="race", y="arms_category", c = "r")
plt.title("arm category per race", fontsize = 15)
plt.xticks(rotation=65)
plt.show()
```

Only the blacks used the hand tools, while the whites and Hispanics used almost all other arm category

```python
df.plot(kind = "scatter", x ="arms_category", y ="age", marker = "D", c = "m")
plt.title("Arm category and age", fontsize=13)
plt.xticks(rotation=65)
plt.show()
```

Gun was preffered, among all ages, while handtools was exclusively for young adults aged around 20.
