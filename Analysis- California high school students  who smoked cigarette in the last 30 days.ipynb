# VISUALIZATION

```python
df.plot()
plt.show()
```

```python
df.plot(kind = "scatter", x = "Strata", y = "Year", c="r")
plt.show()
```

```python
df.plot(kind = "scatter", x = "Denominator", y = "Year", c= "g")
plt.show()
```

The largest sample size was during the 2015-2016 year, while the least was during the 2001-2001

```python
df.plot(kind = "scatter", x = "Rate", y="Year")
plt.show()
```

The rate of smokers reduced  gradually from 2001 to 2016, from 20 in 2001 to 6.0 in 2016

```python
df.plot(kind = "scatter", x = "Lower 95% CI", y="Year", marker = "*", c ="m")
plt.show()
```

5 %chances of being wrong reduced gradually from 2001 to 2016

```python
df.plot(kind = "scatter", x = "Denominator", y="Rate", marker = "o", c ="k")
plt.show()
```

The Rate of smoking reduced as the number of denominator increased. Rate and denominator are inverely proportional

```python
df.plot(kind = "scatter", x = "Lower 95% CI", y="Rate", marker = "o", c ="r")
plt.show()
```

Rate and Lower 95% CI are directly proportional, an increase in Rate of caused an increase in Lower 95%CI
